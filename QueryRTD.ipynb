```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

queryRTD = pd.read_csv('queryRTDFebruary.csv') #funzione per leggere il file csv
queryRTD #per vedere il contenuto del DataFrame
```

```python
# raggruppamento per query (quante volte ogni query appare)
query_counts = queryRTD['Query'].value_counts()

# prime N ricerche piÃ¹ frequenti
top_N = 15
top_queries = query_counts.head(top_N)

# dimensione del grafico
plt.figure(figsize=(10, 6))

# creazione grafico a barre
top_queries.plot(kind='bar', color='skyblue')

# etichette e titolo
plt.xlabel('Query')
plt.ylabel('Numero di Volte')
plt.title('Ricerche piÃ¹ effettuate')

# rotazione delle etichette sull'asse x
plt.xticks(rotation=45, ha='right')

# mostra il grafico
plt.show()
```

```python
# Creazione nuovo DataFrame con query e quante volte appare la query
query_counts_df = pd.DataFrame({'Query': query_counts.index, 'Numero di volte': query_counts.values})

query_counts_df
```
